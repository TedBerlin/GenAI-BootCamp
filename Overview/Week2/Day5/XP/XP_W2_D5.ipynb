5. Differences Among States in Profitability

```python
state_profit_margin = df.groupby('State')[['Sales', 'Profit']].sum()
state_profit_margin['Profit_Margin'] = (state_profit_margin['Profit'] / state_profit_margin['Sales']) * 100
state_profit_margin_sorted = state_profit_margin.sort_values(by='Profit_Margin')

plt.figure(figsize=(14,6))
sns.barplot(x=state_profit_margin_sorted['Profit_Margin'], y=state_profit_margin_sorted.index, palette='coolwarm')
plt.title("Profit Margin by State (%)")
plt.xlabel("Profit Margin (%)")
plt.ylabel("State")
plt.tight_layout()
plt.show()
```

# Observation

10 States get a negative margin, espcially Ohio and Colorado - around -20%. Whereas on the opposite only 3 states are under +20% margin and 12 states are above 30%. Best practices should be retrieved to be given to the states in bad shape.

6. Pareto Principle (80/20) Applied to Customers and Profit

```python
customer_profit = df.groupby('Customer Name')['Profit'].sum().sort_values(ascending=False)
cumulative_profit = customer_profit.cumsum() / customer_profit.sum()

plt.figure(figsize=(12,5))
cumulative_profit.plot(drawstyle='steps-post')
plt.axhline(y=0.8, color='r', linestyle='--')
plt.title("Cumulative Profit by Customer")
plt.xlabel("Customer Rank")
plt.ylabel("Cumulative Profit Contribution")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# How many customers make up 80% of profit?

pareto_cutoff = cumulative_profit[cumulative_profit <= 0.8].count()
total_customers = customer_profit.count()
percentage = (pareto_cutoff / total_customers) * 100
print(f"{percentage:.2f}% of customers contribute to 80% of the profit.")
```

9. Cumulative Curve for Sales by Customer (Pareto Principle)

```python
customer_sales = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False)
cumulative_sales = customer_sales.cumsum() / customer_sales.sum()

plt.figure(figsize=(12,5))
cumulative_sales.plot(drawstyle='steps-post')
plt.axhline(y=0.8, color='r', linestyle='--')
plt.title("Cumulative Sales by Customer")
plt.xlabel("Customer Rank")
plt.ylabel("Cumulative Sales Contribution")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Customer contribution threshold

pareto_cutoff_sales = cumulative_sales[cumulative_sales <= 0.8].count()
total_customers_sales = customer_sales.count()
percentage_sales = (pareto_cutoff_sales / total_customers_sales) * 100
print(f"{percentage_sales:.2f}% of customers contribute to 80% of the sales.")
```

🌟 Exercise 1: Understanding Data Visualization

Explain why data visualization is important in data analysis.


>> Pain Point:

When working with raw data (like tables, CSV files, or JSON), it becomes very difficult to quickly spot trends, outliers, or correlations. Raw data is hard to interpret directly. Spreadsheets and tables don’t reveal insights easily.


A spreadsheet with 10,000 rows doesn’t “speak” to humans. We’re not naturally equipped to detect patterns in columns of numbers.

>> How Data Visualization Solves It:

Data visualization converts abstract data into visual representations (charts, histograms, heatmaps, etc.) that:

    * make trends immediately visible (e.g., growth, seasonality),

    * enable easy comparison across groups or categories,

    * help communicate results clearly to non-technical audiences,

    * assist in spotting errors or outliers in the dataset.

In short, visualization bridges the gap between raw data and human intuition.


It transforms data into meaningful visuals that highlight patterns, trends, outliers, and relationships, making data more accessible and actionable.


>> Future Trends in Data Visualization

    * Interactive Visualizations: Tools like Plotly, Tableau, and Power BI allow users to interact with data (e.g., zoom, filter, hover). This will continue growing, especially for dashboards and real-time analytics.

    * AI-Powered Insights: Visual tools are starting to auto-suggest charts and highlight anomalies or trends using machine learning.

    * Natural Language Interfaces: Users will increasingly use tools that convert questions asked in plain English into dynamic visualizations (e.g., “Show me sales by region”).

    * Immersive Visualization: AR/VR will make it possible to visualize data in 3D environments, especially for geospatial or highly dimensional datasets.

    * Data Storytelling: Visualization will increasingly be used not just to show data, but to tell compelling, narrative-driven stories for business and science.


What Is the Purpose of a Line Graph in Data Visualization?


>> Pain Point:

When tracking a variable over time (e.g., daily temperature, monthly revenue, user growth), it’s hard to judge the overall trend or compare multiple series by just looking at numbers. In fact, Difficult to follow the evolution of values over time using raw numbers.


>> How a Line Graph Solves It:

A line graph:

    * displays continuous values over time or a logical sequence,

    * clearly shows trends (rising, falling, stable),

    * makes it easy to detect spikes, drops, cycles, or anomalies,

    * is great for comparing multiple time series side by side.

A line graph makes trends over time visually clear, highlighting changes, comparisons, and anomalies in a simple way.


>> Future Trends for Line Graphs and Time-Series Visualization

    * Dynamic & Real-Time Plots: Used for monitoring (IoT, finance, health), these plots update live as data streams in.

    * Multi-dimensional Time Series: Future tools will better support overlaying multiple variables, dimensions, and uncertainty bands (e.g., confidence intervals).

    * Enhanced Forecasting Integration: Line graphs will integrate more with predictive models, making it easier to visualize future trends alongside historical data.

    * Customization & Personalization: Tools will adapt visualizations automatically based on the user’s role, intent, or context (e.g., executive vs. data scientist views).



🌟 Exercise 2: Creating a Line Plot for Temperature Variation

```python
# Librairy import

import matplotlib.pyplot as plt
```

```python
# Data definition
# Weekdayus

days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Associated temperatures (en °F)
temperatures = [72, 74, 76, 80, 82, 78, 75]
```

```python
# Graphic creation 

plt.plot(days, temperatures, marker='o', linestyle='-', color='royalblue')

# Axis labels add-up
plt.xlabel('Day')  # x Axis 
plt.ylabel('Temperature (°F)')  # y Axis

# Titel add-up
plt.title('Temperature Variation Over a Week')

# Grid display
plt.grid(True, linestyle='--', alpha=0.6)

# Graphic display
plt.show()
```

🌟 Exercise 3: Visualizing Monthly Sales with a Bar Chart

```python
import matplotlib.pyplot as plt

# Étape 1 : Data definition
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']
sales = [5000, 5500, 6200, 7000, 7500, 7900, 6700, 6100, 7000, 8000, 8200, 7800]

# Étape 2 : Bar chart setup
plt.bar(months, sales, color='royalblue')

# Étape 3 : Add-up of the axis labels
plt.xlabel('Month')
plt.ylabel('Sales Amount ($)')

# Étape 4 : Add-up of a titel to the graphic
plt.title('Monthly Sales for Retail Store')

# Étape 5 : Graphic display
plt.show()
```

🌟 Exercise 5: Comparing Anxiety Levels Across Different Genders

```python
# Import of librairies 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# .CSV uploading

df = pd.read_csv("Student Mental health.csv")
```

```python
# .CSV exploration

# Afficher les premières lignes pour explorer le contenu
print(df.head())
```

```python
# Informations structurelles sur le DataFrame

print("\nℹ️ Dataset information:")
print(df.info())
```

```python
# Statistiques descriptives

print("\nDescriptive statistics:")
print(df.describe())
```

```python
# Vérification des colonnes disponibles

print("\nAvailable columns list:")
print(df.columns)
```

```python
# 4. Nettoyage de base : suppression des lignes où le genre ou l’anxiété n’est pas renseigné
df_clean = df[['Do you have Anxiety?', 'Choose your gender']].dropna()
```

```python
# 5. Créer un graphique de barres avec Seaborn
plt.figure(figsize=(8, 5))
sns.countplot(
    data=df_clean,
    x='Choose your gender',
    hue='Do you have Anxiety?',
    palette='Set2'
)

# 6. Personnaliser le graphique
plt.title('Comparison of Anxiety Levels Across Genders', fontsize=14)
plt.xlabel('Gender')
plt.ylabel('Number of Students')
plt.legend(title='Has Anxiety?')
plt.xticks(rotation=0)
plt.tight_layout()



# 7. Afficher le graphique
plt.show()
```

🌟 Exercise 6: Exploring the Relationship Between Age and Panic Attacks

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# .CSV uploading

df = pd.read_csv("Student Mental health.csv")

# Afficher les premières lignes pour explorer le contenu
print(df.head())
```

```python
# Informations structurelles sur le DataFrame

print("\nℹ️ Dataset information:")
print(df.info())
```

```python
# Statistiques descriptives

print("\nDescriptive statistics:")
print(df.describe())
```

```python
# Vérification des colonnes disponibles

print("\nAvailable columns list:")
print(df.columns)
```

```python
# Renommer les colonnes pour plus de lisibilité
df.columns = df.columns.str.strip()
```

```python
# Sélectionner les colonnes pertinentes
df_plot = df[['Age', 'Do you have Panic attack?']].dropna()
```

```python
# Convertir les réponses en valeurs numériques : Yes → 1, No → 0
df_plot['PanicAttackBinary'] = df_plot['Do you have Panic attack?'].map({'Yes': 1, 'No': 0})
```

```python

# Créer un histogramme de la distribution de l'âge en fonction de la présence d'attaques de panique

# La couleur peut différencier les groupes

sns.histplot(data=df, x='Age', hue='Panic_Attack_Numeric', multiple='stack', palette='Set2')

# Ajout d'un titre pour clarifier le graphique
plt.title('Distribution of Age by Panic Attack Occurrence')

# Ajouter des labels pour plus de clarté
plt.xlabel('Age')
plt.ylabel('Count')
# Afficher le graphique

plt.show()
```

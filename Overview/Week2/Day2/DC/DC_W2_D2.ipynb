```python
# Étape 4 : Charger et afficher une image

from PIL import Image
import matplotlib.pyplot as plt

import os
```

```python
from scipy.ndimage import rotate
import numpy as np

```

```python
# Load an image using PIL
image_path = '0010.png'
image = Image.open("0010.png")

# Display the original image using matplotlib
plt.imshow(image)
```

Étape 5 : Rotation de l’image de 30 degrés

```python
from scipy.ndimage import rotate
import numpy as np

# Convertir l'image PIL en array numpy
image_array = np.array("0010.png")

# Fonction pour faire une rotation de 30 degrés
def rotate_image_30_degrees(image):
  return rotate(image, 30, reshape=False)

rotated_image = rotate_image_30_degrees(image)

# Affichage
plt.imshow(rotated_image.astype(np.uint8))
plt.title("Image tournée de 30°")
plt.axis("off")
plt.show()

```

Étape 6 : Miroir et retournement vertical


```python
from PIL import ImageOps

# Miroir horizontal
mirrored_image = ImageOps.mirror(image)

# Retournement vertical
flipped_image = ImageOps.flip(image)

# Affichage côte à côte
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].imshow(mirrored_image)
axes[0].set_title("Image miroir")
axes[0].axis("off")

axes[1].imshow(flipped_image)
axes[1].set_title("Image retournée verticalement")
axes[1].axis("off")

plt.show()
```

Étape 7 : Zoom (redimensionner à 1.2x)

```python
# Dimensions originales
width, height = image.size

# Nouvelle taille (zoom 1.2x)
zoomed_image = image.resize(
    (int(width * 1.2), int(height * 1.2))
)

# Affichage
plt.imshow(zoomed_image)
plt.title("Image zoomée (1.2x)")
plt.axis("off")
plt.show()
```

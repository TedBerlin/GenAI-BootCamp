```python
# Exercise 1 : Cats

class Cat:
    def __init__(self, cat_name, cat_age):
        self.name = cat_name
        self.age = cat_age

# Cat objects cration
cat1 = Cat("Ursula", 3)
cat2 = Cat("Bologne", 5)
cat3 = Cat("Felix", 1)

# Finding the oldest cat
def find_oldest_cat(*cats):
    oldest_cat = cats[0]
    for cat in cats:
        if cat.age > oldest_cat.age:
            oldest_cat = cat
    return oldest_cat

# The oldest cat display
oldest = find_oldest_cat(cat1, cat2, cat3)
print(f"The oldest cat is {oldest.name}, and is {oldest.age} years old.")
```

```python
# Exercise 2 : Dogs

# Dog Class and Objects / Methods creation

class Dog:
    def __init__(self, name, height):
        self.name = name
        self.height = height
    
    def bark(self):
        print(f"{self.name} goes woof!")
    
    def jump(self):
        jump_height = self.height * 2
        print(f"{self.name} jumps {jump_height} cm high!")
    
    def __gt__(self, other_dog):
        return self.height > other_dog.height
    
    def __lt__(self, other_dog):
        return self.height < other_dog.height
    
    def __eq__(self, other_dog):
        return self.height == other_dog.height

# Dog objects
davids_dog = Dog("Rex", 50)
sarahs_dog = Dog("Teacup", 30)

# Details and call methods
print(f"David's dog: {davids_dog.name}, {davids_dog.height}cm")
davids_dog.bark()
davids_dog.jump()

print(f"\nSarah's dog: {sarahs_dog.name}, {sarahs_dog.height}cm")
sarahs_dog.bark()
sarahs_dog.jump()

# Dog sizes comparison
print("\nSize comparison:")
if davids_dog > sarahs_dog:
    print(f"{davids_dog.name} is taller than {sarahs_dog.name}")
elif davids_dog < sarahs_dog:
    print(f"{sarahs_dog.name} is taller than {davids_dog.name}")
else:
    print(f"Both dogs are the same height")
```

```python
# Exercise 3 : Whoâ€™s the song producer?


class Song:
    def __init__(self, lyrics):
        self.lyrics = lyrics
    
    def sing_me_a_song(self):
        for line in self.lyrics:
            print(line)

# Example usage
stairway = Song([
    "There's a lady who's sure",
    "all that glitters is gold",
    "and she's buying a stairway to heaven"
])

# Print the lyrics
stairway.sing_me_a_song()

```

```python
# Exercise 4 :  Afternoon at the Zoo


class Zoo:
    def __init__(self, zoo_name):
        self.name = zoo_name
        self.animals = []
        self.animal_groups = {}
    
    def add_animal(self, new_animal): # Animal Add-up if not present
        if new_animal not in self.animals:
            self.animals.append(new_animal)
            print(f"Added {new_animal} to {self.name}")
        else:
            print(f"{new_animal} already exists in {self.name}")
    
    def get_animals(self):
        print(f"\nAnimals in {self.name}:") # All animals display
        for animal in self.animals:
            print(f"- {animal}")
    
    def sell_animal(self, animal_sold):
        if animal_sold in self.animals:
            self.animals.remove(animal_sold) #Animal removal if existing
            print(f"\nSold {animal_sold} from {self.name}")
        else:
            print(f"\n{animal_sold} not found in {self.name}")
    
    def sort_animals(self): # Alphabetically sorting
        self.animals.sort()
        self.animal_groups = {}
        
        for animal in self.animals:
            first_letter = animal[0].upper()
            if first_letter not in self.animal_groups:
                self.animal_groups[first_letter] = []
            self.animal_groups[first_letter].append(animal)
        
        return self.animal_groups
    
    def get_groups(self): # Group display
        if not self.animal_groups:
            self.sort_animals()
            
        print(f"\nAnimal groups in {self.name}:")
        for letter, animals in self.animal_groups.items():
            print(f"{letter}: {animals}")


# Zoo instance
cirque_du_soleil = Zoo("Cirque du Soleil")

# Zoo methods
cirque_du_soleil.add_animal("Giraffe")
cirque_du_soleil.add_animal("Bear")
cirque_du_soleil.add_animal("Baboon")
cirque_du_soleil.add_animal("Cougar")
cirque_du_soleil.add_animal("Cat")
cirque_du_soleil.add_animal("Zebra")
cirque_du_soleil.add_animal("Lion")

cirque_du_soleil.get_animals()

cirque_du_soleil.sell_animal("Bear")
cirque_du_soleil.get_animals()

cirque_du_soleil.sort_animals()
cirque_du_soleil.get_groups()
```

Daily Challenge: NumPy, Pandas, and Matplotlib Integration

1. Data preparation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Temperatures dataset for 10 cities / 12 months

```python
np.random.seed(17)  # Reproductibility

cities = ['Paris', 'New York', 'Tokyo', 'Sydney', 'Moscou', 
          'London', 'Rio', 'Bejin', 'Dubai', 'Toronto']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'Mai', 'June', 
        'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Temperatures generation (-5°C to 35°C)
temperatures = np.random.uniform(low=-5, high=35, size=(10, 12))
```

```python
# Dataset creation

df = pd.DataFrame(
    data=np.round(temperatures, 1),  # Arrondi à 1 décimale
    index=cities,
    columns=months
)
print("Generated Dataset:")
display(df.head(5))
```

2. Data analysis

```python
# Average yearly temperature by city

df['Yearly_Mean'] = df.mean(axis=1)
print("\nYearly Mean Temperatures:")
display(df[['Yearly_Mean']].sort_values('Yearly_Mean', ascending=False))
```

```python
hottest_city = df['Yearly_Mean'].idxmax()
coolest_city = df['Yearly_Mean'].idxmin()
print(f"\nHottest city: {hottest_city} ({df['Yearly_Mean'].max():.1f}°C)")
print(f"Coolest city: {coolest_city} ({df['Yearly_Mean'].min():.1f}°C)")
```

3. Data visualisation

```python
# Monthly trends (Top 3 cities)



plt.figure(figsize=(12, 6))
top_3 = df['Yearly_Mean'].nlargest(3).index
for ville in top_3:
    plt.plot(months, df.loc[ville, months], label=ville, marker='o')

plt.title('Monthly temperatures (Top 3 cities)')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
```

```python
# Heatmap des Températures

plt.figure(figsize=(12, 6))
sns.heatmap(df[months], annot=True, fmt=".1f", cmap="coolwarm", linewidths=0.5)
plt.title('Temperature Heatmap by City/Months')
plt.show()
```

The Power of Data Analysis in Modern Decision-Making
1. What is Data Analysis?
Data analysis is the systematic process of inspecting, cleaning, transforming, and interpreting raw data to extract meaningful insights, identify patterns, and support decision-making. It involves various techniques—from basic statistical summaries to advanced machine learning—to turn unstructured or structured data into actionable knowledge 7.

In practice, data analysis can be:

Descriptive (What happened?) – Summarizing past trends (e.g., sales reports).

Diagnostic (Why did it happen?) – Investigating causes (e.g., why hospital readmissions increased).

Predictive (What will happen?) – Forecasting future trends (e.g., disease outbreaks).

Prescriptive (What should we do?) – Recommending actions (e.g., optimizing supply chains) 17.

2. Why is Data Analysis Important Today?
In our data-driven world, analysis is crucial for:

Informed Decision-Making – Replacing guesswork with evidence-based strategies (e.g., Netflix’s recommendation engine improves retention by 93%) 9.

Operational Efficiency – Identifying inefficiencies (e.g., hospitals reducing wait times via patient flow analytics) 5.

Competitive Advantage – Companies like Amazon use dynamic pricing (2.5M daily adjustments) to maximize profits 6.

Risk Mitigation – Banks employ AI-driven fraud detection (e.g., UOB Bank cuts risk assessment time from 18 hours to minutes) 9.

Without data analysis, organizations risk falling behind in agility, customer satisfaction, and innovation 47.

Exercise 4: Exploring Data Types

```python
import pandas as pd

def describe_dataset(df, name=""):
    """Generate a global report on a dataset"""
    print(f"=== Dataset report {name} ===")
    print(f"Shape: {df.shape} (lignes, colonnes)")
    
    # Analyse des types
    print("\nData types:")
    type_map = {
        'object': 'Qualitative (nominal)',
        'category': 'Qualitative (nominal/ordinal)',
        'int64': 'Quantitative (discrete)',
        'float64': 'Quantitative (continuous)',
        'datetime64': 'Quantitative (temporal)'
    }
    
    type_analysis = pd.DataFrame({
        'Colonne': df.columns,
        'Type Python': df.dtypes.values,
        'Type Statistique': [type_map.get(str(dtype), "?") for dtype in df.dtypes]
    })
    display(type_analysis)
    
    # Statistics - basics
    print("\nData overview:")
    display(df.head(3))
    
    print("\nDescriptive statistics")
    display(df.describe(include='all'))
    
    print("\nMissing Values:")
    print(df.isna().sum())

# Use example
sleep_df = pd.read_csv('Time Americans Spend Sleeping.csv')
describe_dataset(sleep_df, "Time Americans Spend Sleeping")

clean_df = pd.read_csv('clean_dataset.csv')
describe_dataset(clean_df, "Clean Dataset")

mental_df = pd.read_csv('Mental health Depression disorder Data.csv')
describe_dataset(mental_df, "Mental Health Data")
```

Exercise 3: Identifying Data Types

```python
import pandas as pd

def classify_data_types(df, dataset_name):
    """Classifies each column as quantitative or qualitative with reasoning"""
    print(f"\n=== Data Type Classification for {dataset_name} ===")
    
    type_map = {
        'object': 'qualitative',
        'category': 'qualitative',
        'bool': 'qualitative',
        'int64': 'quantitative',
        'float64': 'quantitative',
        'datetime64': 'quantitative'
    }
    
    results = []
    
    for column in df.columns:
        dtype = str(df[column].dtype)
        data_type = type_map.get(dtype, 'unknown')
        
        # Custom reasoning
        if column.lower() in ['age', 'minutes', 'price']:
            data_type = 'quantitative'
            reason = "Numerical value allowing mathematical operations"
        elif column.lower() in ['gender', 'condition', 'status']:
            data_type = 'qualitative'
            reason = "Categorical description without numerical meaning"
        elif df[column].nunique() < 10 and dtype == 'int64':
            data_type = 'qualitative'
            reason = "Integer values representing categories (low cardinality)"
        else:
            reason = f"Automatically classified as {data_type} based on dtype"
            
        results.append([column, dtype, data_type, reason])
    
    return pd.DataFrame(results, columns=['Column', 'Dtype', 'Type', 'Reasoning'])

# Load datasets
sleep_df = pd.read_csv('Time Americans Spend Sleeping.csv')
clean_df = pd.read_csv('clean_dataset.csv')
mental_df = pd.read_csv('Mental health Depression disorder Data.csv')

# Each dataset analysis
print("=== DATA TYPE ANALYSIS ===")
display(classify_data_types(sleep_df, "Sleeping Data"))
display(classify_data_types(clean_df, "Medical Data"))
display(classify_data_types(mental_df, "Mental Health Data"))
```

>> Explanations

Key Features:

1. Automatic Classification:

    Numeric dtypes → Quantitative

    Text/bool dtypes → Qualitative

    Special handling for date-like columns


2. Custom Reasoning:

    Identifies categorical numbers (e.g., 1=Male, 2=Female)

    Recognizes common quantitative fields (age, price, etc.)

    Handles edge cases (IDs, binary variables)


3. Visual Presentation:

    Clear tabular output

    Dataset-specific reporting

    Human-readable explanations

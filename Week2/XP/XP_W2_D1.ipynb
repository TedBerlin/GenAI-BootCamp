Exercise 2: Transformation Exercise



1. Série d'articles de blog sur des voyages

>> Méthode de Structuration : NLP + Extraction d'Entités

= Approche : =

Analyse de Texte (NLP) :

    * Tokenisation et Lemmatisation pour isoler les mots-clés.

    * Reconnaissance d'Entités (NER) pour extraire :

        ° Lieux (villes, pays, monuments)

        ° Dates (durée du voyage)

        ° Activités (randonnée, visite, etc.)

    * Analyse de Sentiment pour évaluer l’expérience (positif/négatif/neutre).


Format Structuré (Exemple) :

Blog_ID	Lieu	Activité	Durée	Sentiment	Mots-Clés
1	Paris, France	Visite Musée	5 jours	Positif	Louvre, gastronomie

Outils Recommandés :

    * SpaCy ou NLTK pour le NLP

    * BERT pour une analyse sémantique avancée



2. Enregistrements Audio d’Appels Client

>> Méthode de Structuration : Transcription + Analyse Thématique

= Approche : =

Transcription Automatique (STT) :

    * Convertir l’audio en texte via ASR (Automatic Speech Recognition).

    * Exemple : Google Speech-to-Text ou Whisper (OpenAI).


Extraction d’Informations :

    * Catégorisation (réclamation, question, feedback).

    * Détection d’Intention (ex : "annuler un abonnement").

    * Émotion (tonalité : frustré, satisfait).


Format Structuré (Exemple) :

Call_ID	Transcription	Catégorie	Intention	Émotion
101	"Je veux annuler mon forfait mobile"	Réclamation	Résiliation contrat	Frustration

Outils Recommandés :

    * Whisper (transcription)

    * Dialogflow (analyse d’intention)
    

3. Notes Manuscrites d’une Séance de Brainstorming
Méthode de Structuration : OCR + Réseau de Neurones
Approche :

Reconnaissance d’Écriture (OCR) :

Google Vision AI ou Tesseract pour convertir l’image en texte.

Organisation des Idées :

Clustering (regroupement par thème : marketing, produit, etc.).

Priorisation (mots comme "urgent" ou "idée clé").

Format Structuré (Exemple) :

Note_ID	Texte Extrait	Thème	Priorité	Auteur
1	"Lancer une app mobile"	Développement	Haute	Jean
Outils Recommandés :

Tesseract (OCR)

K-Means (clustering non supervisé)

4. Tutoriel Vidéo de Cuisine
Méthode de Structuration : Vision par Ordinateur + NLP
Approche :

Analyse Vidéo (CV) :

Détection d’Objets (ingrédients, ustensiles).

Reconnaissance d’Actions (émincer, cuire).

Transcription Audio :

Extraire les étapes de la recette.

Structuration Temporelle :

Découpage en étapes avec timestamps.

Format Structuré (Exemple) :

Étape	Ingrédients	Action	Durée	Ustensile
1	Oignon, Huile	Émincer	2 min	Couteau
Outils Recommandés :

OpenCV (détection d’objets)

YOLO (modèle temps réel)

FFmpeg (découpage vidéo)

Synthèse des Méthodes
Source	Technologie Clé	Format Final	Cas d’Usage
Blog posts	NLP + NER	Table Lieux/Activitiés	Analyse touristique
Audio calls	STT + NLP	Base de données clients	Service client automatisé
Notes manuscrites	OCR + Clustering	Matrice idées/thèmes	Gestion de projet
Vidéo tutorielle	Computer Vision + ASR	Recette structurée	Assistant culinaire (IA)

Exercise 5 : Export a dataframe to excel format and JSON format.

```python
import pandas as pd

# Create a simple DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Paris', 'London']
}

df = pd.DataFrame(data)

# Export to Excel format
df.to_excel("sample_data.xlsx", index=False)  # index=False to avoid writing row numbers

# Export to JSON format
df.to_json("sample_data.json", orient="records", indent=4)

# Display the DataFrame
print(df)
```

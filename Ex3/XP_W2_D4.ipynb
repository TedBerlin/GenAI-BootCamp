```python
class Pets():
    def __init__(self, animals):
        self.animals = animals

    def walk(self):
        for animal in self.animals:
            print(animal.walk())

class Cat():
    is_lazy = True

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def walk(self):
        return f'{self.name} is just walking around'

class Bengal(Cat):
    def sing(self, sounds):
        return f'{sounds}'
    
class Chartreux(Cat):
    def sing(self, sounds):
        return f'{sounds}'
    
    def meow (self):
        return f'{self.name} is meowing loudly'


class Siamese(Cat):
    def __init__(self, name, age, color_points="seal point"):
        super().__init__(name, age)
        self.color_points = color_points  # Siamese attribute
        
    def be_aloof(self):
        return f'{self.name} is being really aloof and dignified'
    
    def sing(self, sounds):
        return f'{sounds} in a high-pitched Siamese voice'
    


#  list of cat instances
bengal_obj = Bengal("Tiger", 3)
chartreux_obj = Chartreux("Misty", 5)
siamese_obj = Siamese("Luna", 2, "blue point")

all_cats = [bengal_obj, chartreux_obj, siamese_obj]


#  Pets instance

sara_pets = Pets(all_cats)

# Walk

sara_pets.walk()
```

```python
# Exercise 3: Dogs Domesticated


class Dog:
    def __init__(self, name, age, weight):
        # Attributes coding
        self.name = name
        self.age = age
        self.weight = weight


    def bark(self):
        # Bark message
        return f"{self.name} is barking"

    def run_speed(self):
        # Run speed
        return self.weight / self.age * 10

    def fight(self, other_dog):
        # The winner

        my_power = self.run_speed() * self.weight
        other_power = other_dog.run_speed() * other_dog.weight
        
        if my_power > other_power:
            return f"{self.name} won the fight"
        elif my_power < other_power:
            return f"{other_dog.name} won the fight"
        else:
            return "It's a tie!"

# Dog instances creation
#... your code here
dog1 = Dog("Rex", 3, 15)
dog2 = Dog("Max", 5, 20)
dog3 = Dog("Buddy", 4, 18)


# Dog methods
print(dog1.bark())
print(dog2.run_speed())
print(dog1.fight(dog2))
```

```python
# Exercise 3: Test PetDog Methods

import random
from dog import Dog  # Cf dog.py file

class PetDog(Dog):
    def __init__(self, name, age, weight):
        super().__init__(name, age, weight)
        self.trained = False

    def train(self):
        print(self.bark())
        self.trained = True

    def play(self, *dogs):
        dog_names = [dog.name for dog in dogs]
        print(f"{self.name}, {', '.join(dog_names)} all play together")

    def do_a_trick(self):
        if self.trained:
            tricks = [
                "does a barrel roll",
                "stands on his back legs",
                "shakes your hand",
                "plays dead"
            ]
            print(f"{self.name} {random.choice(tricks)}")
        else:
            print(f"{self.name} isn't trained yet!")

# Testing the PetDog class
if __name__ == "__main__":
    # Create some dogs
    fido = PetDog("Fido", 2, 10)
    buddy = PetDog("Buddy", 3, 15)
    max = PetDog("Max", 4, 20)
    
    # Test methods
    print("\n--- Training ---")
    fido.train()
    
    print("\n--- Playing ---")
    fido.play(buddy, max)
    
    print("\n--- Tricks ---")
    fido.do_a_trick()
    buddy.do_a_trick()  # Not trained yet
    
    print("\n--- Training Buddy ---")
    buddy.train()
    buddy.do_a_trick()
```

```python
# Exercise 4: Family and Person Classes

class Person: # Class creation
    def __init__(self, first_name, age):
        self.first_name = first_name
        self.age = age
        self.last_name = ""  
    
    def is_18(self):
        return self.age >= 18


class Family: # Family class creation
    def __init__(self, last_name):
        self.last_name = last_name
        self.members = []  # Person stockage
    
    def born(self, first_name, age):
        new_person = Person(first_name, age)
        new_person.last_name = self.last_name  
        self.members.append(new_person)
        print(f"Welcome {first_name} {self.last_name} into the family!")
    
    def check_majority(self, first_name):
        for person in self.members:
            if person.first_name == first_name:
                if person.is_18():
                    parents = "Jane and John"  # Noms des parents par défaut
                    print(f"You are over 18, your parents {parents} accept that you will go out with your friends.")
                else:
                    print("Sorry, you are not allowed to go out with your friends.")
                return
        print(f"{first_name} has not been found in the family.")
    
    def family_presentation(self):
        print(f"\nFamily {self.last_name} :")
        for member in self.members:
            print(f"- {member.first_name}, {member.age} years old")


# Test du système
if __name__ == "__main__":
    # Family creation
    family_doe = Family("Doe")
    
    # Members Add-up
    family_doe.born("Alice", 20)
    family_doe.born("Bob", 15)
    family_doe.born("Charlie", 10)
    
    # Majority checking
    print("\nGo out Verification:")
    family_doe.check_majority("Alice")  # Major
    family_doe.check_majority("Bob")    # Minor
    family_doe.check_majority("Eve")    # No existing
    
    # Family display
    family_doe.family_presentation()
```
